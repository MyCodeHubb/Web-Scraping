```python
#hitory development and planning
#1
questions=[]

result=requests.get("https://www.gkseries.com/general-knowledge/indian-economy/history-of-development-and-planning/history-of-development-and-planning")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
#2 
result_2=requests.get("https://www.gkseries.com/general-knowledge/indian-economy/history-of-development-and-planning/00002-history-of-development-and-planning")
soup = bs4.BeautifulSoup(result_2.text,"lxml")
for item in soup.select('.mcq'):
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)

    
#3-8
base_url='https://www.gkseries.com/general-knowledge/indian-economy/history-of-development-and-planning/mcqs-on-history-of-development-and-planning-{}'
for n in range(3,9):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup= bs4.BeautifulSoup(res.text,'lxml')
    
    for item in soup.select('.mcq'):
        a_list=item.text.split()
        a_list.remove('View')
        a_list.remove('Answer')
        a_list.remove('Comment')
        new_string=" ".join(a_list)
        
        questions.append(new_string)

```

```python
schema={"Topic":"Indian Economy","Subject":"History Development and Planning","data":questions}
json_object=json.dumps(schema,indent=4)
with open("historydevandplan.json",'w') as json_file:
    json_file.write(json_object)
```

```python
json_file.close()
```
