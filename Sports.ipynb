```python
import requests
import bs4
import json
```

```python
#football
questions=[]
#1
result=requests.get("https://www.gkseries.com/general-knowledge/sports/00001-world-cup-football-quiz")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
    
#2    
result=requests.get("https://www.gkseries.com/general-knowledge/sports/football/00002-latest-gk-quiz")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
#3    
result=requests.get("https://www.gkseries.com/general-knowledge/sports/football/00003-football-gk")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
    
schema={"Topic":"Sports","Subject":"Football","data":questions}
json_object=json.dumps(schema,indent=4)
with open("football.json",'w') as json_file:
    json_file.write(json_object)
    
json_file.close()
```
