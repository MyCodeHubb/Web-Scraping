```python
import requests
import bs4
import json
```

```python
#constitutional
#1
questions=[]

result=requests.get("https://www.gkseries.com/general-knowledge/indian-polity/constitutional-development-in-india/constitutional-development-in-india")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
#2 
result_2=requests.get("https://www.gkseries.com/general-knowledge/indian-polity/constitutional-development-in-india/00002-constitutional-development-in-india")
soup = bs4.BeautifulSoup(result_2.text,"lxml")
for item in soup.select('.mcq'):
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)

    
#3-25
base_url='https://www.gkseries.com/general-knowledge/indian-polity/constitutional-development-in-india/mcqs-on-constitutional-development-in-india-{}'
for n in range(3,25):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup= bs4.BeautifulSoup(res.text,'lxml')
    
    for item in soup.select('.mcq'):
        a_list=item.text.split()
        a_list.remove('View')
        a_list.remove('Answer')
        a_list.remove('Comment')
        new_string=" ".join(a_list)
        
        questions.append(new_string)
        
schema={"Topic":"Indian Polity","Subject":"Constitutional Development","data":questions}
json_object = json.dumps(schema,indent=4)
with open("constitutionaldev.json",'w') as json_file:
    json_file.write(json_object)        

json_file.close()    

```
