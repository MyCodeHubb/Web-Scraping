```python
import requests
```

```python
import bs4
```

```python
#indian history
#1
questions=[]

result=requests.get("https://www.gkseries.com/general-knowledge/indian-history/sources-of-indian-history/sources-of-indian-history")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
    
    
#2
result=requests.get("https://www.gkseries.com/general-knowledge/indian-history/sources-of-indian-history/00002-sources-of-indian-history")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
        

        

#3-12
base_url='https://www.gkseries.com/general-knowledge/indian-history/sources-of-indian-history/gk-msqs-on-sources-of-indian-history-{}'
for n in range(3,13):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup= bs4.BeautifulSoup(res.text,'lxml')
    
    for item in soup.select('.mcq'):
        a_list=item.text.split()
        a_list.remove('View')
        a_list.remove('Answer')
        a_list.remove('Comment')
        new_string=" ".join(a_list)
        
        questions.append(new_string)
       

    
```

```python
schema={"Topic":"Indian History","Subject":"Sources of Indian History","data":questions}
json_object=json.dumps(schema,indent=4)
with open("indianhistory.json",'w') as json_file:
    json_file.write(json_object)
```

```python
json_file.close()
```
