```python
#animation
questions=[]
#1
result=requests.get("https://www.gkseries.com/animation-and-editing/questions-and-answers-on-animation-and-editing")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    
    new_string=" ".join(a_list)
        
    questions.append(new_string)
    
#2-5

base_url='https://www.gkseries.com/animation-and-editing/questions-and-answers-on-animation-and-editing-{}'
for n in range(2,6):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup= bs4.BeautifulSoup(res.text,'lxml')
    
    for item in soup.select('.mcq'):
        a_list=item.text.split()
        new_string=" ".join(a_list)
        questions.append(new_string)
        

schema={"Topic":"Science","Subject":"Animation","data":questions}
json_object = json.dumps(schema,indent=4)
with open("animation.json",'w') as json_file:
    json_file.write(json_object)
    
json_file.close()


```
