```python
import requests
import bs4
import json
```

```python
#noble prize
#1
questions=[]

result=requests.get("https://www.gkseries.com/questions-and-answers-on-awards-honours-prizes/nobel-prize/general-knowledge-questions-and-answers-on-nobel-prize")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
    
    
#2
result=requests.get("https://www.gkseries.com/questions-and-answers-on-awards-honours-prizes/nobel-prize/gk-on-nobel-prize-2")
soup = bs4.BeautifulSoup(result.text,"lxml")
for item in soup.select('.mcq'):
    
    a_list=item.text.split()
    a_list.remove('View')
    a_list.remove('Answer')
    a_list.remove('Comment')
    new_string=" ".join(a_list)
        
    questions.append(new_string)
        

        

#3-6
base_url='https://www.gkseries.com/questions-and-answers-on-awards-honours-prizes/nobel-prize/general-knowledge-questions-and-answers-on-nobel-prize-{}'
for n in range(3,7):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup= bs4.BeautifulSoup(res.text,'lxml')
    
    for item in soup.select('.mcq'):
        a_list=item.text.split()
        a_list.remove('View')
        a_list.remove('Answer')
        a_list.remove('Comment')
        new_string=" ".join(a_list)
        
        questions.append(new_string)
       

    
```

```python
schema={"Topic":"Awards/Honours/Prizes","Subject":"Nobel Prize","data":questions}
with open("nobelprize.json",'w') as json_file:
    json.dump(schema,json_file)
    
json_file.close()
```
